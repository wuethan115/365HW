```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
```

```python
# Q2.1

np.random.seed(365)
n = 100
mu_true = 10
sigma_true = 1
data = np.random.normal(loc = mu_true, scale = sigma_true, size = n)

with pm.Model() as model:
    # Prior Theta
    theta = pm.Normal("theta", mu = 0, sigma = 10)

    # Prior Tau
    tau = pm.Gamma("tau", alpha = 1, beta = 1)

    # Likelihood
    y_obs = pm.Normal("y_obs", mu = theta, sigma = 1/pm.math.sqrt(tau),
                      observed = data)

    # Inference
    trace = pm.sample(2000, tune = 1000, return_inferencedata = True,
                      cores = 2, target_accept = 0.9)

# Theta Posterior
theta_posterior = trace.posterior["theta"].values.flatten()
theta_hpd = az.hdi(theta_posterior, hdi_prob = 0.95)
print(f"95% Credible Interval for θ: {theta_hpd}")

# Tau Posterior
tau_posterior = trace.posterior["tau"].values.flatten()
tau_hpd = az.hdi(tau_posterior, hdi_prob = 0.95)
print(f"95% Credible Interval for τ: {tau_hpd}")

# Diagnostics
fig, axes = plt.subplots(2, 2, figsize = (10, 6))
az.plot_trace(trace, compact = False, axes = axes)

plt.tight_layout()
plt.show()

az.summary(trace)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

## Q3: Markov algorithm of slice sampling

Slice sampling is a Markov Chain Monte Carlo method that generates samples from a target distribution by iteratively defining and sampling from a "slice" beneath its probability density function. Instead of directly sampling from the target distribution, the algorithm first introduces a supplementary variable $y\sim\text{Uniform}(0, f(x))$, sampled uniformly between $0$ and $f(x)$. This defines a horizontal slice where $f(x)>y$.

The "slice_f_at_y" function determines the interval $[x_l,x_r]$ where $f(x)$ is greater than $y$, ensuring that endpoints satisfy $f(a)<y$ and $f(b)<y$.  A new $x$ is then sampled uniformly from this interval. If the proposed $x$ is within the valid region $\left(f(x)>y\right)$, it is accepted -- if the proposed $x$ is not within the valid region $\left(f(x)\leq y\right)$, the interval is shrunk recursively using the "slice_f_at_y_" function until a valid sample is obtained.

The main loop iterates this process to generate a Markov Chain of samples, which eventually converges to the target distribution.

## Q3: Replacing Metropolis-Hastings

In a Metropolis within Gibbs algorithm, Gibbs sampling is used to update each variable sequentially from its full conditional distribution. However, when the full conditional is only known up to a normalizing constant, direct sampling is more difficult.

In this situation, instead of using a Metropolis-Hastings step to propose and accept/reject new values, we can use slice sampling to replace the Metropolis-Hastings step. This is done by adaptively defining a sampling region that allows us to avoid using a proposal distribution. Instead of proposing a new value and computing an acceptance probability, slice sampling introduces the supplementary variable $y$ as defined previously. The slice sampling algorithm then runs until a valid sample is obtained.

The curve being sampled beneath in this case is the un-normalized full conditional probability distribution -- this is typically a posterior distribution that is known up to a normalizing constant. The initial value is arbitrary and set by the user at the start of the algorithm.



```python
# Implementation of Slice Sampling as a replacement for MH

import numpy as np
from scipy import stats

def gibbs_with_slice_sampling(full_conditional, x_init, num_iters,
                              x_range = (0, 1), num_grid = 100):
    samples = np.zeros(num_iters)
    x = x_init

    x_grid = np.linspace(x_range[0], x_range[1], num_grid)

    for i in range(num_iters):
        # Sample slice height y from Uniform(0, f(x))
        y = np.random.uniform(0, full_conditional(x))

        # Determine the interval where f(x) > y
        valid_x = x_grid[full_conditional(x_grid) > y]
        if len(valid_x) == 0:
            raise ValueError("No valid slice interval found.")

        x_l, x_r = valid_x[0], valid_x[-1]

        # Sample new x uniformly from slice interval
        x = np.random.uniform(x_l, x_r)
        samples[i] = x
    return samples
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
